```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the car crashes dataset
crashes = sns.load_dataset('car_crashes')

# Plot the total number of car crashes by state
sns.barplot(x='total', y='abbrev', data=crashes.sort_values('total', ascending=False))
plt.xlabel('Total Car Crashes')
plt.ylabel('State Abbreviation')
plt.title('Total Car Crashes by State')
plt.show()

# Plot the number of speeding car crashes by state
sns.barplot(x='speeding', y='abbrev', data=crashes.sort_values('speeding', ascending=False))
plt.xlabel('Speeding Car Crashes')
plt.ylabel('State Abbreviation')
plt.title('Speeding Car Crashes by State')
plt.show()
```
